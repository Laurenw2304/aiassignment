```python
import pandas as pd
df = pd.read_csv('Marketing_Design_Dataset.csv-assignment.csv')
df.head(4)
```

# New section

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Campaign_Budget', data=df, palette='Blues')
plt.title('Campaign Budget by Region')
plt.xlabel('Region')
plt.ylabel('Campaign Budget ($)')
plt.show()
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Ad_Click_Rate', y='Conversion_Rate', data=df, hue='Platform', style='Campaign_Type')
plt.title('Ad Click Rate vs. Conversion Rate')
plt.xlabel('Ad Click Rate (%)')
plt.ylabel('Conversion Rate (%)')
plt.show()
plt.figure(figsize=(10, 6))
sns.boxplot(x='Target_Audience', y='Conversion_Rate', data=df, palette='coolwarm')
plt.title('Conversion Rate by Target Audience')
plt.xlabel('Target Audience')
plt.ylabel('Conversion Rate (%)')
plt.xticks(rotation=45)
plt.show()
```
